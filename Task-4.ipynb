```python
desired_order=['customer_id','name','first_name','last_name','email','phone_number','address','state_code','country_code','registration_date','loyalty_status','customer_tier','order_id','order_date','order_amount','order_status','product_category']
```

```python
cursor.execute('''
create table unified_customer_view(
customer_id int,
name varchar(100),
first_name varchar(50),
last_name varchar(50),
email varchar(100),
phone_number varchar(50),
address nvarchar(250),
state_code varchar(25),
country_code varchar(25),
registration_date date,
loyalty_status varchar(25),
customer_tier int,
order_id varchar(200),
order_date date,
order_amount float,
order_status varchar(50),
product_category varchar(100)
)''')
conn.commit()
```

```python
for row in unified_view.itertuples():
    cursor.execute('''
    insert into unified_customer_view(customer_id, name, first_name, last_name, email,
       phone_number, address, state_code, country_code,
       registration_date, loyalty_status, customer_tier, order_id,
       order_date, order_amount, order_status, product_category)
       values(?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)
       ''',
                   row.customer_id,
                   row.name,
                   row.first_name,
                   row.last_name,
                   row.email,
                   row.phone_number,
                   row.address,
                   row.state_code,
                   row.country_code,
                   row.registration_date,
                   row.loyalty_status,
                   row.customer_tier,
                   row.order_id,
                   row.order_date,
                   row.order_amount,
                   row.order_status,
                   row.product_category
                  )
conn.commit()
```
